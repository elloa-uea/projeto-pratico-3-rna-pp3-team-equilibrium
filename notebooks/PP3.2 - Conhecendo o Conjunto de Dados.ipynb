**Atributo Elevação (Elevation)**

```python
elevation_arr = data['Elevation'].to_numpy()

fig, axs = plt.subplots(2)
fig.set_size_inches(9, 10)

axs[0].hist(elevation_arr, bins=500)
axs[0].set_title('Histograma do atributo Elevação')
axs[0].set_xlabel('Elevação(m)', fontsize=14)
axs[0].set_ylabel('Quantidade de exemplos', fontsize=14)

flierprops = dict(marker='o', markerfacecolor='r', markersize=8, linestyle='none', markeredgecolor='g')
axs[1].boxplot([elevation_arr],vert=False, labels=[''], flierprops=flierprops)
axs[1].set_title('Boxplot do atributo Elevação')
axs[1].set_xlabel('Elevação(m)', fontsize=14)

fig.tight_layout(pad=3.0)

plt.show()

print("Algumas informações estatísticas")

print("Média: " + 
      str(np.round(data.Elevation.describe()[1], 4)) + "m")

print("Desvio padrão: " + 
      str(np.round(data.Elevation.describe()[2], 4)) + "m")

print("Menor valor: " + 
      str(np.round(data.Elevation.describe()[3], 4)) + "m")

print("Valor do primeiro quartil: " + 
      str(np.round(data.Elevation.describe()[4], 4)) + "m")

print("Mediana: " + 
      str(np.round(data.Elevation.describe()[5], 4)) + "m")

print("Valor do terceiro quartil: " + 
      str(np.round(data.Elevation.describe()[6], 4)) + "m")

print("Intervalo interquartil: " +
      str(np.round((data.Elevation.describe()[6] - data.Elevation.describe()[4]), 4)) + "m")

print("Maior valor: " + 
      str(np.round(data.Elevation.describe()[7], 4)) + "m")
```
